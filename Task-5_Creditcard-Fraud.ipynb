```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('creditcard.csv')
df = pd.DataFrame(data)
df.head()
```

```python
#Training the model and splitting the data
X = df.drop('Class',axis=1)
y=data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
print(f'''X_train:{X_train.shape}
      X_test:{X_test.shape}
      y_train:{y_train.shape}
      y_test:{y_test.shape}''')
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
```

```python
randomForestModel = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(
        n_estimators=100,  # Start with 100 trees
        max_depth=20,      # Restrict depth of trees
        n_jobs=-1
    ))
])
randomForestModel.fit(X_train, y_train)
```

```python
y_pred_logics = randomForestModel.predict(X_test)
y_pred_logics 
```

```python
Report = classification_report(y_test,y_pred_logics)
print(Report)
```
